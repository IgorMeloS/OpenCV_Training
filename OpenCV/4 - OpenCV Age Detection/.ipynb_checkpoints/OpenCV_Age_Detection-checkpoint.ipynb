# Age Detection with OpenCV

This notebook presents a simple way to make age prediction using OpenCV. There two forms to predict age, training your model from scratch or consider a pre-trained model. For this example, we consider a pre-trained model and we use the dnn module from OpenCV.

The age prediction requires two steps:

- Face detection (we extract the region of interest - ROI)
- Age prediction (classification inside the ROI).

We consider the pre-trained models (face and age recognition) and their weights from the CAFFE framework. 
The original work for the age prediction can be found [here](https://talhassner.github.io/home/projects/cnn_agegender/CVPR2015_CNN_AgeGenderEstimation.pdf). If you want to train your model from scratch, visit [here]().

We use four pictures by three Brazilian singers to illustrate this example.

## Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import os
```

## Building the age detection with pre-trained models

We consider a classification model to the age prediction. So, is more conveniently to separate the age in certains ranges.

```python
# Define the list of age range
AGE_RAN = ["(0-2)", "(4-6)", "(8-12)", "(15-20)", "(25-32)", "(38-43)", "(48-53)", "(60-100)"]
```

**Loading the face detection model**

We consider a caffe pre-trained model for the face detection and age prediction. If you want to learn more about CAFFE, I suggest the read in their [Website](https://caffe.berkeleyvision.org/) and [Github](https://github.com/BVLC/caffe/).

**OpenCV dnn module**

The OpenCV dnn module offers to us many functions to work with Deep Neural NetWork. The better way to know all application from this modules is the read of your module, [here](https://docs.opencv.org/4.5.0/d6/d0f/group__dnn.html).

```python
prototxt = os.path.sep.join(["deploy.prototxt"]) # read the model
weights = os.path.sep.join(["res10_300x300_ssd_iter_140000.caffemodel"]) # read the weights
faceNet = cv.dnn.readNet(prototxt, weights) # building the network with dnn module
```

**Loading the age detector model**

```python
prototxtage = os.path.sep.join(["age_deploy.prototxt"])
weightsage = os.path.sep.join(["age_net.caffemodel"])
ageNet = cv.dnn.readNet(prototxtage, weightsage)
```

**Loading the images from the disk and making a list of images**

```python
dataset =  os.path.abspath("dataset") # the path for the dataset folder
image_list = os.listdir(dataset) # the list with all images inside the dataset folder
```

**Main loop for face detection and age predictions**

To proceed with the age prediction some operations are required when we use the dnn module.

- Load the image from disk
- Grab the image coordinates
- Use the blob function
    - [blobFromImage](https://docs.opencv.org/master/d6/d0f/group__dnn.html#ga29f34df9376379a603acd8df581ac8d7) creates 4-dimensional blob from image. Optionally resizes and crops image from center, subtract mean values, scales values by scalefactor, swap Blue and Red channels.
- Input the returned image from the blob in the network
- Make face detection with the pre-trained network
- To filter the low confidence in the prediction array
- Extract the region of interest, it means grab the coordinates of the face detection
- To make the age prediction

```python
count = 1 # counter to build the pyplot display
plt.figure(figsize=(20, 20)) # creating the figure

# Starting the main loop to make prediction for each image inside the image list
for j in image_list:
    
    img = dataset + "/" + j # setting te image
    image = cv.imread(img) # reading the image with opencv
    (h, w) = image.shape[:2] # extacting the image coordinates
    
    blob = cv.dnn.blobFromImage(cv.resize(image, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))
    
    # Communicating the blob through the network and obtain the face detections
    faceNet.setInput(blob)
    detections = faceNet.forward()
    
    # Loop for the prediction on each image
    for i in range(0, detections.shape[2]):
        
        # Extract the confidence for all predictions
        confidence = detections[0, 0, i, 2]
        
        # Condition to eliminate the low confidences
        if confidence > 0.5:
            
            # coordinates of the detected faces
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (X, Y, dX, dY) = box.astype("int")
            
            # Defining the region of interest to each detected face and applying the blob to them
            face = image[Y:dY, X:dX]
            faceBlob = cv.dnn.blobFromImage(face, 1.0, (227, 227), (78.4263377603, 87.7689143744, 114.895847746),
                                             swapRB=False)
            
            # Making the age prediction
            ageNet.setInput(faceBlob)
            preds = ageNet.forward() # age predictions
            i = preds[0].argmax() # grab the great confidence
            age = AGE_RAN[i] # grab the age range
            ageConfidence = preds[0][i] # putting the confidence
            
            # Display the predicted age to the terminal
            text = "{}: {:.2f}%".format(age, ageConfidence * 100)
            
            # Draw the bounding box of the face along with the associated predicted age
            y = Y - 10 if Y - 10 > 10 else Y + 10
            cv.rectangle(image, (X, Y), (dX, dY), (0, 0, 255), 2)
            cv.putText(image, text, (X, y), cv.FONT_HERSHEY_SIMPLEX, 0.45, (0, 0, 255), 2)
        
    ax = plt.subplot(3, 2, count)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(image_list[count-1])
    plt.axis("off")
    count += 1

plt.savefig("age_prediction.jpg")
plt.show()
```

## Conclusion

The results are not satisfactory. The problem to use pre-trained model is the our limitation in respect to the weights and the data balance. The considered dataset for this model is the [Adience](https://talhassner.github.io/home/projects/Adience/Adience-data.html#agegender) dataset and, from the original work, we can see that for certain  age range the results show some bias. The solution for this issue is to train with other images and consider class balance.
