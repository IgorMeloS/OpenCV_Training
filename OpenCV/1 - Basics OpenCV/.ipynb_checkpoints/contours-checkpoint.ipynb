# Contours

The contours might be an useful tool for computer vision tasks. For example, we can count the number of objects in a certain image, once we have the contours. We can also crop the image, selecting the desired region delimited by the contour and, many others applicability. To find contours and draw them, we must consider some techniques as convert to gray scale, histogram equalization, blurring, edge detection and thresholding. 

In this example, we cover

- Find contours
- Draw contours

The image is the Brazilian team of 1980, in the button football version.

## Importing libraries

```python
from __future__ import print_function
import numpy as np
import cv2
import imutils
import matplotlib.pyplot as plt
```

## Load image from disk

```python
img = cv2.imread("images/but_foot.jpg")
```

```python
plt.figure(figsize=(15,10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Countours

In this part, we proceed with the elementary actions to find the contours. We start converting the image into gray scale, then we apply histogram equalization. For the blurring, we consider the function cv2.GaussinaBlur, wiht a kernel of $(9,9)$, these values was found by experimentation. The next step, we consider the function cv2.Canny, for the edge detection. The minimal threshold is $45$ and the maximum is $216$, we also consider the penalty L2gradient. To delimit the colors into a binary scale, we consider the function cv2.threshold, using the the function cv2.THRESH_BINARY and, the minimal pixel is $200$ and the maximum is $255$.

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # converting into gray scale
gray = gray.astype(np.uint8)
eq = cv2.equalizeHist(gray) # histogram equalization
blurred = cv2.GaussianBlur(eq, (9,9), 0) # blurring
edged = cv2.Canny(blurred, 45, 216, L2gradient = True, apertureSize = 3) # edge detection
(T, thresh) = cv2.threshold(edged, 200, 255, cv2.THRESH_BINARY) # thresholding
```

**visualizing the transformations**

```python
plt.figure(figsize=(15,15))
plt.subplot(3,2,1)
plt.title("Gray scale")
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.subplot(3,2,2)
plt.title("Histogram Equalization")
plt.imshow(cv2.cvtColor(eq, cv2.COLOR_BGR2RGB))
plt.subplot(3,2,3)
plt.title("Blurred")
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.subplot(3,2,4)
plt.title("Edge detection")
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
plt.subplot(3,2,5)
plt.title("Threshold")
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
plt.show()
```

**Finding contours**

**from OpenCV website**

**What are contours?**

Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

- For better accuracy, use binary images. So before finding contours, apply threshold or canny edge detection.
- Since OpenCV 3.2, findContours() no longer modifies the source image.
- In OpenCV, finding contours is like finding white object from black background. So remember, object to be found should be white and background should be black.

The function to find contours is **cv2.findContous(src, mode, method)**. To better understand the mode and method, [see](https://docs.opencv.org/4.5.2/d3/dc0/group__imgproc__shape.html#gadf1ad6a0b82947fa1fe3c3d497f260e0). We consider the image given by the thresholding, the mode is RETR_EXTERNAL and, the method is CHAIN_APPROX_SIMPLE. The mode allows us to grab just the highers contours. The function returns the contours. We use also grab_contours from imutils, to put the contours in a single python list.

```python
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
```

```python
# Print the number of found contours
print("I count {} objects in this image".format(len(cnts)))
```

**Drawing contours**

**from OpenCV wwebsite**

To draw the contours, cv.drawContours function is used. It can also be used to draw any shape provided you have its boundary points. Its first argument is source image, second argument is the contours which should be passed as a Python list, third argument is index of contours (useful when drawing individual contour. To draw all contours, pass -1) and remaining arguments are color, thickness etc.

First of all, we make a copy of the original image and then, we apply the function, the arguments are the copy of the original image, the found contours, $-1$ to draw all contours, the tuple to set the color and $2$ is the thickness of the line we are drawing.

```python
coins = img.copy()
coins = cv2.drawContours(coins, cnts, -1, (0, 255, 0), 2)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(cv2.cvtColor(coins, cv2.COLOR_BGR2RGB))
plt.show()
```
